```python
import nltk
from nltk.corpus import brown
from nltk.util import ngrams
from collections import defaultdict, Counter
import string

# Download dataset
nltk.download('brown')

# Load Brown corpus
words = brown.words()

# Preprocessing
words = [word.lower() for word in words if word not in string.punctuation]

# Set N value (change this to 3, 4, 5, etc.)
N = 3   # Trigram model

# Build N-gram model
ngram_model = defaultdict(Counter)

for gram in ngrams(words, N):
    context = gram[:-1]   # first N-1 words
    next_word = gram[-1]  # last word
    ngram_model[context][next_word] += 1

# Prediction function
def predict_next_word_ngram(context_words):
    context = tuple(context_words)

    if context in ngram_model:
        return ngram_model[context].most_common(1)[0][0]
    else:
        return "No prediction available"

# Example usage
context = ["in", "the"]   # N-1 words required
print("Context:", context)
print("Predicted next word:", predict_next_word_ngram(context))
```

```python
import nltk
from nltk.corpus import brown
from collections import defaultdict, Counter

nltk.download('brown')

# Load and preprocess
sentences = brown.sents()
cleaned_sentences = []

for sent in sentences:
    cleaned = []
    for w in sent:
        if w.isalpha():
            cleaned.append(w.lower())
    cleaned_sentences.append(cleaned)

# -------- USER INPUT FOR N --------
n = int(input("Enter N value for N-Gram (e.g., 2 for Bigram, 3 for Trigram): "))

# Build N-Gram Model
ngram_model = defaultdict(Counter)

for sent in cleaned_sentences:
    for i in range(len(sent) - n + 1):
        context = tuple(sent[i:i+n-1])
        next_word = sent[i+n-1]
        ngram_model[context][next_word] += 1

print("\nN-Gram Model Built Successfully!")

# -------- USER INPUT FOR CONTEXT --------
context_input = input(f"Enter {n-1} words as context: ")
context_words = tuple(context_input.lower().split())

# Validate context length
if len(context_words) != n-1:
    print(f"Error: Please enter exactly {n-1} words!")
else:
    # Predict next word
    if context_words in ngram_model:
        print("Count of possible words:", len(ngram_model[context_words]))
        predicted_word = ngram_model[context_words].most_common(1)[0][0]
    else:
        predicted_word = "No prediction available for this context"

    print("\nContext:", context_input)
    print("Predicted Next Word (N-Gram):", predicted_word)
```
