```python
import nltk
from nltk.corpus import brown
from collections import defaultdict, Counter
import string

# Downloads
nltk.download('brown')
nltk.download('universal_tagset')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

# Load tagged sentences
tagged_sentences = brown.tagged_sents(tagset='universal')

# Transition and emission counts
transition = defaultdict(Counter)
emission = defaultdict(Counter)

# Build HMM parameters (REMOVE PUNCTUATION)
for sent in tagged_sentences:
    for i in range(len(sent) - 1):
        curr_word, curr_tag = sent[i]
        next_tag = sent[i + 1][1]

        curr_word = curr_word.lower()

        # Ignore punctuation tokens
        if curr_word.isalpha():
            transition[curr_tag][next_tag] += 1
            emission[curr_tag][curr_word] += 1

# Function to get POS tag of a word
def get_pos_tag(word):
    tagged = nltk.pos_tag([word], tagset='universal')
    return tagged[0][1]

# Prediction function
def predict_next_word_hmm(last_word):
    last_word = last_word.lower()

    if not last_word.isalpha():
        return "Invalid input"

    last_tag = get_pos_tag(last_word)

    if last_tag not in transition:
        return "No prediction available"

    # Predict next POS tag
    next_tag = transition[last_tag].most_common(1)[0][0]

    # Predict most frequent alphabetic word
    for word, _ in emission[next_tag].most_common():
        if word.isalpha():
            return word

    return "No prediction available"

# -------- USER INPUT --------
user_input = input("Enter a word: ").strip()

predicted_word = predict_next_word_hmm(user_input)

print("Input word:", user_input)
print("Predicted next word:", predicted_word)
```

```python
import nltk
from nltk.corpus import brown
from collections import defaultdict, Counter

nltk.download('brown')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

# Load tagged sentences
tagged_sents = brown.tagged_sents()

transition = defaultdict(Counter)
emission = defaultdict(Counter)

# Train HMM parameters
for sent in tagged_sents:
    prev_tag = "<s>"
    for word, tag in sent:
        word = word.lower()
        transition[prev_tag][tag] += 1
        emission[tag][word] += 1
        prev_tag = tag

# Input word
input_word = "the"

# POS tag of input word
tagged_word = nltk.pos_tag([input_word])
current_tag = tagged_word[0][1]

# Predict next tag
if current_tag in transition:
    next_tag = transition[current_tag].most_common(1)[0][0]
else:
    next_tag = None

# Predict next word from emission
if next_tag in emission:
    print("Count of possible words:", len(emission[next_tag]))
    predicted_word = emission[next_tag].most_common(1)[0][0]
else:
    predicted_word = "No prediction available"

print("Input Word:", input_word)
#print("Current POS Tag:", current_tag)
#print("Predicted Next POS Tag:", next_tag)
print("Predicted Next Word (HMM):", predicted_word)
```
